# Tarefa 1

### 1. Idade media das mães e dos pais por município (coluna munResNome)


```python
sinascpaiemae = sinasc_raw.groupby('munResNome')[['IDADEMAE', 'IDADEPAI']].mean()
sinascpaiemae
```

### 2. Peso médio dos bebes por sexo que nasceram no dia do seu aniversário por faixas de escolaridade mae
Ex: Você, aluna(o), nasceu no dia 10/01, então você precisa filtrar o conjunto de dados nessa data e calcular o peso médio dos bebês de cada sexo por faixa de escolaridade da mãe.

```python
sinascpeso = sinasc_raw.groupby(['SEXO', 'ESCMAE'])['PESO'].mean()
sinascpeso
```

### 3. Qual o municipio que nasceu menos bebe em 2019?
    - qual a idade media, maxima, minima das maes nesse municipio?
    - qual a idade media, maxima, minima dos pais nesse municipio?

```python
sinasc_raw.groupby('munResNome')['DTNASC'].count().sort_values(ascending=True)[0:6].to_frame()
```

```python
sinasc_raw.groupby('munResNome').aggregate({
    'DTNASC': [('Contagem', 'count')],
    'IDADEMAE': [('Média','mean'), ('Máxima','max'), ('Mínima', 'min')]
}).sort_values(by=('DTNASC','Contagem'), ascending=True)[:6]
```

```python
sinascidade1 = sinasc_raw.groupby('munResNome')['IDADEMAE'].mean()
sinascidade1
```

```python
sinasc_raw.groupby('munResNome').aggregate({
    'DTNASC': [('Contagem', 'count')],
    'IDADEPAI': [('Média','mean'), ('Máxima','max'), ('Mínima', 'min')]
}).sort_values(by=('DTNASC','Contagem'), ascending=True)[:6]
```

### 4. Qual o municipio que nasceu mais bebe no mês de março?
    - qual a quantidade de filhos vivos media, maxima, minima nesse municipio?
    - qual a idade media, maxima, minima dos pais nesse municipio?



```python
SINASCMARCO = sinasc_raw[sinasc_raw['DTNASC'].str.startswith('2019-03')]
SINASCMARCO.groupby('munResNome').aggregate({
    'DTNASC': [('Contagem', 'count')],
    'QTDFILVIVO': [('Média','mean'), ('Máxima','max'), ('Mínima', 'min')]
}).sort_values(by=('DTNASC','Contagem'), ascending=False)[:6]
```

```python
sinasc_raw.groupby('munResNome').aggregate({
    'DTNASC': [('Contagem', 'count')],
    'IDADEMAE': [('Média','mean'), ('Máxima','max'), ('Mínima', 'min')]
}).sort_values(by=('DTNASC','Contagem'), ascending=False)[:6]
```

- qual a idade media, maxima, minima dos pais nesse municipio?

```python
sinasc_raw.groupby('munResNome').aggregate({
    'DTNASC': [('Contagem', 'count')],
    'IDADEPAI': [('Média','mean'), ('Máxima','max'), ('Mínima', 'min')]
}).sort_values(by=('DTNASC','Contagem'), ascending=False)[:6]
```
